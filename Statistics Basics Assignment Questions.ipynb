```python
#12 ans)
import numpy as np

# Example dataset
X = np.array([10, 12, 15, 18, 20])
Y = np.array([20, 25, 30, 35, 40])

# Calculate means
mean_X = np.mean(X)
mean_Y = np.mean(Y)

# Covariance
covariance = np.cov(X, Y, ddof=1)[0, 1]

# Correlation
correlation = np.corrcoef(X, Y)[0, 1]

covariance, correlation
```
